```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
df = pd.read_csv("farmer.csv")
```

```python
newdf = pd.DataFrame(df["text"])
```

```python
print("not all the lowecases :",newdf.iloc[45,0])
```

```python
newdf['text'] = newdf['text'].str.lower()
print("all the lowercase :",newdf.iloc[45,0])
```

```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    text_nopunt="".join([i  for i in text if i not in string.punctuation])
    return text_nopunt
```

```python
newdf['text']= newdf['text'].apply(lambda x : remove_punctuation(x))
print("removing puncuation :",newdf.iloc[45,0])
```

```python
newdf['text'] = newdf['text'].str.replace(r'http\S+|www.\S+','')
print("removing link :",newdf.iloc[45,0])
```

```python
print("with numbers in string :",newdf.iloc[57,0])
```

```python
newdf['text'] = newdf['text'].str.replace("\d+","")
print("without numbers in string :",newdf.iloc[57,0])
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
stop
```

```python
print("with stopwords :",newdf.iloc[45,0])
```

```python
newdf['text'] = newdf['text'].apply(lambda x: ' '.join([abc for abc in x.split() if abc not in (stop)]))
print('without any  stopwords :' ,newdf.iloc[45,0])
```

```python
import re 
nltk.download('words') 
words = set(nltk.corpus.words.words()) 
words
```

```python
from nltk.stem import WordNetLemmatizer 
import nltk
nltk.download('wordnet') 
lemmatizer = WordNetLemmatizer()
w_tokenizer = nltk.tokenize.WhitespaceTokenizer() 
```

```python
def lemmatize_text(text):
    return [lemmatizer.lemmatize(i,"v") for i in w_tokenizer.tokenize(text)]
```

```python
newdf['text'] = newdf["text"].apply(lambda x :' '.join(lemmatize_text(x)))
newdf.iloc[991,0]
```

```python
newdf['text'] = (newdf['text'].astype("str").str.rstrip())
newdf.iloc[991,0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
data_cv = cv.fit_transform(newdf["text"])
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names()) 
tdm = data_dtm.transpose()
tdm.head()
```

```python
tdm['freq'] = tdm.sum(axis=1)
tdm.head()
```

```python
tdm.reset_index(inplace=True)
tdm.head()
```

```python
tdm1=tdm[["index","freq"]] 
tdm1.rename(columns = {'index':'Word'}, inplace = True)  
tdm1.sort_values(by='freq',ascending=False,inplace=True)  
tdm1.head()
```

```python
text = " ".join(i for i in newdf["text"])
print ("There are {} words in the combination of all review.".format(len(text)))
```
